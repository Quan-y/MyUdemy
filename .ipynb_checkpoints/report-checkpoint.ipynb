```python
# import necessary package
import pandas as pd
import numpy as np
from scipy import stats
from scipy.sparse import hstack, csr_matrix
import time

import seaborn as sns
import matplotlib.pyplot as plt

from nltk.corpus import stopwords
from wordcloud import WordCloud, STOPWORDS

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn_pandas import DataFrameMapper
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import preprocessing
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import Ridge

from mlxtend.regressor import StackingRegressor
from lightgbm import LGBMRegressor

%matplotlib inline
```

```python
# all functions

# function calculate statistics of price, for EDA
def cal_bar(data, col, percent):
    bar = 1
    rate = 0
    while rate <= percent:
        rate = len(data[col][data[col] <= bar])*1.0/len(data[col])
        bar += 1
    return bar

# splite category_name
def split_cat(text):
    try: return text.split("/")
    except: return ("No Label", "No Label", "No Label")

# analysis the relationship between feature and obj, for EDA
def cause_effect(data, cause, plot = False):
    collect = []
    for i in list(set(data[cause])):
        collect.append(data[data[cause] == i]['price'].describe())
    collect_df = pd.concat(collect, axis = 1)
    collect_df.columns = [cause + str(i) for i in list(set(data[cause]))]
    collect_df.loc['count_rate'] = (collect_df.iloc[0]/sum(collect_df.iloc[0].tolist())).tolist()
    if plot == True:
        f2, ax2 = plt.subplots(figsize = (15, 8))
        sns.boxplot(x = data[cause], y = data['price'], ax = ax2)
    return collect_df

# rmsle
def rmsle(y_true,y_pred):
    assert len(y_true) == len(y_pred)
    return np.square(np.log(y_pred + 1) - np.log(y_true + 1)).mean() ** 0.5

# function for preprocessing txt data
def txt_to_vec(txt, method = 'idf', n_features = 10):
    if type(txt[0]) == dict:
        # model
        vec = DictVectorizer()
        # fit
        return [vec.fit_transform(txt), vec.get_feature_names()]
    elif type(txt[0]) == str:
        if method == 'idf':
            vec = TfidfVectorizer(norm = 'l2', min_df = 13, max_df = 0.99, stop_words = 'english', \
                                  analyzer = 'word', max_features = 50000)
            return [vec.fit_transform(txt), vec.get_feature_names()]
        elif method == 'hash':
            vec = HashingVectorizer(n_features = n_features, stop_words = 'english', analyzer = 'word')
            return [vec.fit_transform(txt)]
    else:
        print 'error'

def filter_str(sen):
    return filter(lambda ch: ch in 'abcdefghijklmnopqrstuvwxyz ABCDEFGHIJKLMNOPQRSTUVWXYZ 0123456789.', sen.lower())
```

## 2. Benchmark Algorithm

```python
# make benchmark dataset
benchmark_data = alldata_adj.drop(['train_id', 'category_name'], axis = 1)

# feature engineering, labelencoder and len
# replace non with 0
benchmark_data.item_description = benchmark_data.item_description.replace('No description yet', '0')
benchmark_data.item_description = benchmark_data.item_description.replace(np.nan, '0')
benchmark_data.brand_name = benchmark_data.brand_name.replace(np.nan, '0')
mapper = DataFrameMapper([
    (['category_1'], LabelEncoder()),
    (['category_2'], LabelEncoder()),
    (['category_3'], LabelEncoder()),
    (['brand_name'], LabelEncoder())
])
benchmark_data.loc[:, ['category_1', 'category_2', 'category_3', 'brand_name']] = mapper.fit_transform(benchmark_data)
benchmark_data['name'] = benchmark_data['name'].map(len)
benchmark_data['item_description'] = benchmark_data['item_description'].map(len)

# modeling
x = benchmark_data.drop(['price'],axis =1)
y = benchmark_data.price
# splite data set into train and test
##############################################################################################
# Step1: train and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.4, random_state = 1)
print "Training set has {} samples.".format(x_train.shape[0])
print "Testing set has {} samples.".format(x_test.shape[0])
# use benchmark model to train it

start = time.time()
# Step2: build model
reg_benchmark = RandomForestRegressor(random_state = 0)
# Step3: training in the train dataset
reg_benchmark.fit(x_train, y_train)
# Step4: predict in the test dataset
y_pred = reg_benchmark.predict(x_test)

end = time.time()
# Step5: evaluation
print 'benchmark model trainset rmsle equals to :',rmsle(y_train, reg_benchmark.predict(x_train))
print 'benchmark model testset rmsle equals to :',rmsle(y_test, y_pred)
print 'the training time is ',end - start
```

```python
# cross validation
# over fitting
# generalization
cross_val_score(reg_benchmark, x, y, cv = 5, scoring = 'neg_mean_squared_error')
```

## 3. Same Algorithm with better feature engineering

```python
# get data for modeling

# handle missing data
model_data = alldata_adj.drop(['train_id', 'category_name'], axis = 1)
model_data.item_description = model_data.item_description.replace('No description yet', '0')
model_data.item_description = model_data.item_description.replace(np.nan, '0')
model_data.brand_name = model_data.brand_name.replace(np.nan, '0')

# lable encode category data
mapper = DataFrameMapper([
    (['category_1'], LabelEncoder()),
    (['category_2'], LabelEncoder()),
    (['category_3'], LabelEncoder()),
    (['brand_name'], LabelEncoder())
])
model_data.loc[:, ['category_1', 'category_2', 'category_3', 'brand_name']] = mapper.fit_transform(model_data)
model_data.head()
```

### 3.1 Tf-idf

```python
# preprocessing txt data

# 1. clean txt, filter unmber and english word, remove symbol
item_description = model_data['item_description'].map(filter_str)
name = model_data['name'].map(filter_str)

# 2. tf-idf
item_description_idf = txt_to_vec(item_description)
name_idf = txt_to_vec(name)
print 'the number of derivatives from item description is ', len(item_description_idf[1])
print 'the number of derivatives from name is ', len(name_idf[1])

# 3. process model_data
price_1 = model_data['price']
model_data_1 = model_data.drop(['name', 'item_description', 'price'],axis=1)
model_data_1 = csr_matrix(model_data_1.values)
model_data_1 = hstack([model_data_1, item_description_idf[0]])
model_data_1 = hstack([model_data_1, name_idf[0]])
```

### 3.2 Truncated SVD

```python
# SVD for reduce dimenstion
# model
svd = TruncatedSVD(n_components = 10, algorithm = 'arpack', tol = 0.01)
# fit
svd.fit(model_data_1.T)
print 'svd explained variance ratio '
print svd.explained_variance_ratio_
print 'svd explained variance ratio sum ', svd.explained_variance_ratio_.sum()

# generate new dataset
new_features = svd.components_
model_data_2 = model_data.drop(['name', 'item_description', 'price'], axis=1)
model_data_2['svd_1'], model_data_2['svd_2'], model_data_2['svd_3'] = \
new_features[0], new_features[1], new_features[2]
```

### 3.3 Rerun Benchmark model with better features

```python
# split the dataset
# this train and test dataset will be used in the following part
x_train, x_test, y_train, y_test = train_test_split(model_data_2, price_1, test_size = 0.4, random_state = 1)
print "Training set has {} samples.".format(x_train.shape[0])
print "Testing set has {} samples.".format(x_test.shape[0])

# re run randomforeast
start = time.time()

reg_1 = RandomForestRegressor(random_state = 0)
reg_1.fit(x_train, y_train)
y_pred_1 = reg_1.predict(x_test)

end = time.time()
print 'benchmark model with better features trainset rmsle equals to :',rmsle(y_train, reg_1.predict(x_train))
print 'benchmark model with better features rmsle equals to :',rmsle(y_test, y_pred_1)
print 'the training time is ',end - start
```

```python
# cross validation
cross_val_score(reg_1, model_data_2, price_1, cv = 5, scoring = 'neg_mean_squared_error')
```

## 5. Same features, same algorithm with better parameters

```python
# adjust parameters of lightgbm(n_estimators and learning rate)
lgb_parameters = {'learning_rate' : [0.1, 0.15],
                  'colsample_bytree': [0.8,0.9],
                  'subsample' : [0.8, 0.9]}
lgb_best = LGBMRegressor(n_jobs = 4, num_leaves = 31,\
                         min_child_samples = 50, n_estimators = 1000
                        )
reg_5 = GridSearchCV(lgb_best, lgb_parameters)
reg_5.fit(x_train, y_train)
y_pred_5 = reg_5.predict(x_test)
print 'Best lightgbm model rmsle equals to :',rmsle(y_test, y_pred_5)
print 'best parameters are: ', reg_5.best_params_
```

```python
start = time.time()
reg_best = LGBMRegressor(n_jobs = 4, num_leaves = 31,\
                         min_child_samples = 50, n_estimators = 1000,\
                         learning_rate = 0.15, colsample_bytree = 0.9,\
                         subsample = 0.9)
reg_best.fit(x_train, y_train)
y_pred_best = reg_best.predict(x_test)
print 'Best LightGBM trainset rmsle equals to :',rmsle(y_train, reg_best.predict(x_train))
print 'Best LightGBM testset rmsle equals to :',rmsle(y_test,y_pred_best)
end = time.time()
print end - start
```

## 6. Ensemble Learning

### 6.1 average each predict results

```python
# try simple to use mean
y_pred_6 = (y_pred_1 + y_pred_2 + y_pred_best + y_pred_4)/4
print 'ensemble model with average results rmsle equals to :',rmsle(y_test,y_pred_6)

# don't use ridge
y_pred_7 = (y_pred_1 + y_pred_2 + y_pred_best)/3
print 'ensemble model with average results(not ridge) rmsle equals to :',rmsle(y_test,y_pred_7)
```

### 6.2 weighted average

```python
# try to add weight
weight = [1.0/0.156, 1.0/0.173, 1.0/0.152, 1.0/0.192]
norm_weight = preprocessing.normalize(weight, norm='l1')[0]
y_pred_8 = norm_weight[0]*y_pred_1 + norm_weight[1]*y_pred_2 + norm_weight[2]*y_pred_best + norm_weight[3]*y_pred_4
print 'ensemble model with weighted results rmsle equals to :',rmsle(y_test,y_pred_8)

# don't use ridge
weight = [1.0/0.156, 1.0/0.173, 1.0/0.152]
norm_weight = preprocessing.normalize(weight, norm='l1')[0]
y_pred_9 = norm_weight[0]*y_pred_1 + norm_weight[1]*y_pred_2 + norm_weight[2]*y_pred_best
print 'ensemble model with weighted results rmsle(not ridge) equals to :',rmsle(y_test,y_pred_9)
```

### 6.3 Stacking

```python
# let's do stacking
stack_1 = StackingRegressor(regressors = [reg_1, reg_2, reg_3],   
                          meta_regressor = reg_best)
stack_1.fit(x_train, y_train)
y_pred_10 = stack_1.predict(x_test)
print 'Stacking model rmsle equals to :',rmsle(y_test,y_pred_10)

# don't use ridge
stack_2 = StackingRegressor(regressors = [reg_1, reg_2],   
                          meta_regressor = reg_best)
stack_2.fit(x_train, y_train)
y_pred_11 = stack_2.predict(x_test)
print 'Stacking model without ridge rmsle equals to :',rmsle(y_test,y_pred_11)
```
